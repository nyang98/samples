```python
#visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

#print the graphs in the notebook
%matplotlib inline

#set seaborn style to white
sns.set_style('white')
```

```python
#plot the age column histogram
ttbill = sns.distplot(shootings.age);

#set labels and titles
ttbill.set(xlabel = 'Age', ylabel = 'Frequency', title = 'Age of Victims')

#take out the right and upper borders
sns.despine()
```

```python
#present the relationship between race and age value
sns.stripplot(x="race", y="age", data = shootings, jitter = True);
```

```python
#create a scatter plot with the race as the y-axis and age as the x-axis, differ the dots by gender
sns.stripplot(x='age', y='race', hue='gender', data = shootings, jitter=True);
```

```python
#create two scatterlpot graphs one for male and another for female, presenting the totalbill value and tip relationship differing by smoker or no smoker
g = sns.FacetGrid(shootings, col = 'race', hue = 'gender')
g.map(plt.scatter, 'age', 'age', alpha =.7)

g.add_legend();
```
