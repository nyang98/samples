```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
```

```python
potle = pd.read_csv(url, sep ='\t')
```

```python
potle.item_price.dtype
```

```python
# clean the item_price column and transform it in a float
prices = [float(value[1:-1]) for value in potle.item_price]
```

```python
prices
```

```python
#reassign the column with the cleaned prices
potle.item_price = prices
```

```python
#delete the duplicates in item name and quantity
potle_filtered= potle.drop_duplicates(['item_name','quantity','choice_description'])
potle_filtered
```

```python
#select only the products with quantity equals to 1
potle_one_prod = potle_filtered[potle_filtered.quantity == 1]
potle_one_prod
```

```python
#lists number of distinct item prices
potle_one_prod[potle_one_prod['item_price']>10].item_name.nunique()
```

```python
potle_one_prod[potle_one_prod['item_price']>10]
```

```python
#sort by the name of the item
potle.sort_values(by='item_name')
```

```python
#what was the quantity of the most expensive item ordered?
potle.sort_values(by='item_price',ascending=False).head(1)
```
